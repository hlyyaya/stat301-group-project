### Data
#### Overview
The selected dataset for this project is the "[World University Rankings 2023](https://www.kaggle.com/datasets/alitaqi000/world-university-rankings-2023/data)," which comprises data on 1,799 universities across 104 countries and regions. This dataset is a comprehensive collection of information regarding university rankings based on various performance indicators.

#### Data Source 
The dataset is publicly accessible and downloadable on Kaggle, where Syed Ali Taqi is listed as the author. The provenance is accredited to [the *Times Higher Education* World University Ranking 2023](https://www.timeshighereducation.com/world-university-rankings/2023/world-ranking). The data was collected via web scraping, then preprocessed and manipulated using Pandas DataFrame and other Python libraries. 
By analyzing over 121 million citations from more than 15.5 million research publications, as well as survey responses from 40,000 scholars globally, the researchers accumulated more than 680,000 data points from more than 2,500 participating institutions that submitted data.

The dataset is expected to be updated annually on Kaggle, with the last update being 6 months ago. The version used for this project was downloaded on February 17, 2024 from Kaggle and is shown below. 

```r
# Read in data:
rankings_data <- read_csv(file = "https://raw.githubusercontent.com/czhang176/stat301_project/main/World%20University%20Rankings%202023.csv")
rankings_data
```

#### Data Characteristics 
- Number of Observations: 2,341
- Number of Variables: 13

Each variable represents a performance indicator that measures an institution’s performance across four areas: teaching, research, knowledge transfer and international outlook.

### Exploratory Data Analysis
#### Cleaning and Wrangling Dataset

```r
# Explore the structure and summary of the dataset
str(rankings_data)
summary(rankings_data)
```

```r
# Replace empty strings, "-", "n/a", and "NaN" with NA
rankings_data[rankings_data == ""] <- NA
rankings_data[rankings_data == "-"] <- NA
rankings_data[rankings_data == "n/a"] <- NA
rankings_data[rankings_data == "NaN"] <- NA

# Convert columns to appropriate data types
rankings_data$`No of student` <- as.integer(rankings_data$`No of student`)
rankings_data$`No of student per staff` <- as.numeric(rankings_data$`No of student per staff`)
rankings_data$`International Student` <- as.numeric(sub("%", "", rankings_data$`International Student`)) / 100
rankings_data$`Female:Male Ratio` <- as.numeric(sapply(strsplit(rankings_data$`Female:Male Ratio`, ":"), function(x) x[1])) / 
                                      (as.numeric(sapply(strsplit(rankings_data$`Female:Male Ratio`, ":"), function(x) x[1])) +
                                       as.numeric(sapply(strsplit(rankings_data$`Female:Male Ratio`, ":"), function(x) x[2])))
rankings_data$`OverAll Score` <- as.numeric(rankings_data$`OverAll Score`)
rankings_data$`Teaching Score` <- as.numeric(rankings_data$`Teaching Score`)
rankings_data$`Research Score` <- as.numeric(rankings_data$`Research Score`)
rankings_data$`Citations Score` <- as.numeric(rankings_data$`Citations Score`)
rankings_data$`Industry Income Score` <- as.numeric(rankings_data$`Industry Income Score`)
rankings_data$`International Outlook Score` <- as.numeric(rankings_data$`International Outlook Score`)

# Drop rows with missing values
rankings_data <- na.omit(rankings_data)

rankings_data
```

```r
# Splitting the data into train and test sets
set.seed(42)
split <- initial_split(rankings_data, prop = 0.8, strata = 'Teaching Score')
train_data <- training(split)
test_data <- testing(split)

# Remove variables that may pose problems
train_data <- select(train_data, -c('University Rank', 'Name of University', 'Location'))
test_data <- select(test_data, -c('University Rank', 'Name of University', 'Location'))

# Define a linear regression model
mlr_model <- linear_reg() %>%
  set_engine("lm") %>%
  set_mode("regression")

# Define the formula for MLR
formula <- as.formula('`Teaching Score` ~ .')

# Train the MLR model
mlr_model <- mlr_model %>%
  fit(formula, data = train_data)

# Predict on the test set
predictions <- predict(mlr_model, new_data = test_data)

# Evaluate model performance
mse <- mean((predictions$.pred - test_data$`Teaching Score`)^2)
mse
```

```r
library(ggplot2)

# Create a data frame for visualization
visualization_data <- data.frame(
  Actual = test_data$`Teaching Score`,
  Predicted = predictions$.pred
)

# Create a scatter plot
ggplot(visualization_data, aes(x = Actual, y = Predicted)) +
  geom_point(color = "blue") +
  geom_abline(slope = 1, intercept = 0, color = "red", linetype = "dashed") +
  labs(
    title = "Actual vs. Predicted Teaching Scores",
    x = "Actual Teaching Score",
    y = "Predicted Teaching Score"
  ) +
  theme_minimal()
```

```r
# Calculate residuals
residuals <- test_data$`Teaching Score` - predictions$.pred

# Plot residuals
plot(residuals, ylab = "Residuals", main = "Residual Plot")
abline(h = 0, col = "red")
```

```r
# Histogram of residuals
hist(residuals, main = "Histogram of Residuals", xlab = "Residuals")

# Shapiro-Wilk test for normality
shapiro.test(residuals)
```

```r
# Calculate mean absolute error (MAE)
mae <- mean(abs(residuals))
cat("Mean Absolute Error (MAE):", mae, "\n")

# Calculate root mean squared error (RMSE)
rmse <- sqrt(mean(residuals^2))
cat("Root Mean Squared Error (RMSE):", rmse, "\n")
```

```r
# Define a mapping between countries and regions
country_region_mapping <- list(
  "United Kingdom" = "Western Europe",
  "United States" = "North America",
  "Switzerland" = "Western Europe",
  "Canada" = "North America",
  "Australia" = "Australia",
  "Singapore" = "Asia",
  "France" = "Western Europe",
  "Sweden" = "Scandinavia",
  "China" = "Asia",
  "Netherlands" = "Western Europe",
  "Japan" = "Asia",
  "Germany" = "Western Europe",
  "South Korea" = "Asia",
  "Belgium" = "Western Europe",
  "Finland" = "Scandinavia",
  "Denmark" = "Scandinavia",
  "Austria" = "Western Europe",
  "Norway" = "Scandinavia",
  "New Zealand" = "Australia",
  "Spain" = "Southern Europe",
  "Italy" = "Southern Europe"
)

# Map countries to regions
rankings_data$Region <- unlist(country_region_mapping[rankings_data$Location])

# Splitting the data into train and test sets
set.seed(42)
split <- initial_split(rankings_data, prop = 0.8, strata = 'Teaching Score')
train_data <- training(split)
test_data <- testing(split)

# Remove variables that may pose problems
train_data <- select(train_data, -c('University Rank', 'Name of University', 'Location'))
test_data <- select(test_data, -c('University Rank', 'Name of University', 'Location'))

# Define a linear regression model
mlr_model <- linear_reg() %>%
  set_engine("lm") %>%
  set_mode("regression")

# Define the formula for MLR
formula <- as.formula('`Teaching Score` ~ .')

# Train the MLR model
mlr_model <- mlr_model %>%
  fit(formula, data = train_data)

# Predict on the test set
predictions <- predict(mlr_model, new_data = test_data)

# Evaluate model performance
mse <- mean((predictions$.pred - test_data$`Teaching Score`)^2)
mse
```

```r
# Create a data frame for visualization
visualization_data <- data.frame(
  Actual = test_data$`Teaching Score`,
  Predicted = predictions$.pred
)

# Create a scatter plot
ggplot(visualization_data, aes(x = Actual, y = Predicted)) +
  geom_point(color = "blue") +
  geom_abline(slope = 1, intercept = 0, color = "red", linetype = "dashed") +
  labs(
    title = "Actual vs. Predicted Teaching Scores",
    x = "Actual Teaching Score",
    y = "Predicted Teaching Score"
  ) +
  theme_minimal()
```

```r
# Aggregate data by region and calculate mean teaching score for each region
region_summary <- rankings_data %>%
  group_by(Region) %>%
  summarise(
    Mean_Teaching_Score = mean(`Teaching Score`, na.rm = TRUE),
    Median_Teaching_Score = median(`Teaching Score`, na.rm = TRUE),
    Min_Teaching_Score = min(`Teaching Score`, na.rm = TRUE),
    Max_Teaching_Score = max(`Teaching Score`, na.rm = TRUE)
  )

# Create a box plot for aggregated teaching scores by region
ggplot(region_summary, aes(x = Region, y = Mean_Teaching_Score, fill = Region)) +
  geom_bar(stat = "identity") +
  labs(
    title = "Distribution of Teaching Scores by Region (Aggregated)",
    x = "Region",
    y = "Mean Teaching Score"
  ) +
  theme_minimal()
```
